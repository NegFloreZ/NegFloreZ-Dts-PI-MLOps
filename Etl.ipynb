```python
# Importar librerias.
import pandas as pd
import numpy as np
```

```python
# Carga de los datasets por plataforma.
df_amazon=pd.read_csv('Datasets/amazon_prime_titles.csv')
df_disney=pd.read_csv('Datasets/disney_plus_titles.csv')
df_hulu=pd.read_csv('Datasets/hulu_titles.csv')
df_netflix=pd.read_csv('Datasets/netflix_titles.csv')
```

```python
# Carga de los datasets ocn los puntajes.
df_ratings1=pd.read_csv('Datasets/ratings/1.csv')
df_ratings2=pd.read_csv('Datasets/ratings/2.csv')
df_ratings3=pd.read_csv('Datasets/ratings/3.csv')
df_ratings4=pd.read_csv('Datasets/ratings/4.csv')
df_ratings5=pd.read_csv('Datasets/ratings/5.csv')
df_ratings6=pd.read_csv('Datasets/ratings/6.csv')
df_ratings7=pd.read_csv('Datasets/ratings/7.csv')
df_ratings8=pd.read_csv('Datasets/ratings/8.csv')
```

```python
# Se agrega la inicial segun la plataforma para complementar el campo id y hacerlo unico.
df_amazon['show_id']=df_amazon['show_id'].map('a{}'.format)
df_amazon['platform']= 'amazon'
df_disney['show_id']=df_disney['show_id'].map('d{}'.format)
df_disney['platform']= 'disney'
df_hulu['show_id']=df_hulu['show_id'].map('h{}'.format)
df_hulu['platform']= 'hulu'
df_netflix['show_id']=df_netflix['show_id'].map('n{}'.format)
df_netflix['platform']= 'netflix'
```

```python
# Union de los datasets para continuar con las transformaciones.
df_stream_platforms = pd.concat([df_amazon,df_disney,df_hulu,df_netflix], ignore_index=True)
df_ratings = pd.concat([df_ratings1,df_ratings2,df_ratings3,df_ratings4,df_ratings5,df_ratings6,df_ratings7,df_ratings8], ignore_index=True)

# Cambio del nombre de las colummnas para ajustarlo al requerimiento.
df_stream_platforms=df_stream_platforms.rename(columns={'show_id':'id'})
df_ratings=df_ratings.rename(columns={'movieId':'id'})
df_ratings=df_ratings.rename(columns={'rating':'score'})
```

```python
# Creacion de la columna score basada en el promedio de los puntajes.
score=df_ratings.groupby(['id'],as_index=False)['score'].mean()
```

```python
# Se agrega la columna score ajustandola a su correspondiente id.
df_stream_platforms=pd.merge(df_stream_platforms,score, how='outer',on='id')
df_stream_platforms
```

```python
# Se reemplazan los valores faltantes del campo rating con la clasificacion G.
df_stream_platforms['rating']=df_stream_platforms['rating'].replace([np.nan],'G')
```

```python
# Funcion querelaiza la conversion de la fecha a un valor numerico:
def fecha_a_numero(dato):
    # Asignar un valor dummie en caso de ser un campo faltante.
    if dato == "nofecha":
        return "1991/1/1"
    else:
    # Manejo de listas para asignar cada parte de la fecha:
        lfecha=[]
        lmes=[]
    # Se recorre le dato para ir extrayendo los campos.
        for i in str(dato):
            lfecha.append(i)

        for i in lfecha:
            if i == " ":
                break
            else:
                lmes.append(i)
        mes=''.join(lmes)

        lfecha=lfecha[len(lmes):]
        laño= lfecha[(len(lfecha)-4):len(lfecha)]
        ldia= lfecha[(len(lfecha)-8):(len(lfecha)-4)]
    # Se filtra por mes para asignar su valor correspondiente.
        if mes == 'January':
            laño.append("/")
            laño.append("1")
            laño.append("/")
        elif mes == 'February':
            laño.append("/")
            laño.append("2")
            laño.append("/")
        elif mes == 'March':
            laño.append("/")
            laño.append("3")
            laño.append("/")
        elif mes == 'April':
            laño.append("/")
            laño.append("4")
            laño.append("/")
        elif mes == 'May':
            laño.append("/")
            laño.append("5")
            laño.append("/")
        elif mes == 'June':
            laño.append("/")
            laño.append("6")
            laño.append("/")
        elif mes == 'July':
            laño.append("/")
            laño.append("7")
            laño.append("/")
        elif mes == 'August':
            laño.append("/")
            laño.append("8")
            laño.append("/")
        elif mes == 'September':
            laño.append("/")
            laño.append("9")
            laño.append("/")
        elif mes == 'October':
            laño.append("/")
            laño.append("10")
            laño.append("/")
        elif mes == 'November':
            laño.append("/")
            laño.append("11")
            laño.append("/")
        else:
            laño.append("/")
            laño.append("12")
            laño.append("/")
       
    # Se recorre el string para determinar el dia.
        for i in ldia:
            if i == ",":
                break
            elif i == " ":
                continue
            else:
                laño.append(i)
        resultado=''.join(laño)
        return resultado
    
```

```python
# Se reemplazan los valores faltantes de la fecha.
df_stream_platforms['date_added']= df_stream_platforms['date_added'].replace(np.nan,'nofecha')
```

```python
# Se modifica el valor en string por un valor numerico.
df_stream_platforms['date_added']= df_stream_platforms['date_added'].map(lambda x:fecha_a_numero(x))
```

```python
# Se hace el cambio de tipo de dato y se formatea segun el requerimiento.
df_stream_platforms['date_added']= pd.to_datetime(df_stream_platforms['date_added'], format='%Y/%m/%d')
```

```python
#Se estandarizan los valores a caracteres en minuscula.
df_stream_platforms['id'] =df_stream_platforms['id'].str.lower()
df_stream_platforms['type'] =df_stream_platforms['type'].str.lower()
df_stream_platforms['director'] =df_stream_platforms['director'].str.lower()
df_stream_platforms['cast'] =df_stream_platforms['cast'].str.lower()
df_stream_platforms['country'] =df_stream_platforms['country'].str.lower()
df_stream_platforms['rating'] =df_stream_platforms['rating'].str.lower()
df_stream_platforms['duration'] =df_stream_platforms['duration'].str.lower()
df_stream_platforms['listed_in'] =df_stream_platforms['listed_in'].str.lower()
df_stream_platforms['description'] =df_stream_platforms['description'].str.lower()

```

```python
# la funcion permite extraer el tipo de duracion en el campo duration.
def conversion_numero_duration(dato):
    if dato == 'NaN':
        return np.nan
    else:
        duration=[]
        numero=[]
        for i in str(dato):
            duration.append(i)

        for i in duration:
            if i == " ":
                break
            else:
                numero.append(i)
        resulta=''.join(numero)
    return resulta
```

```python
# la funcion permite extraer el tiempo de duracion en el campo duration.
def conversion_tipo_duration(dato):
    if dato == 'NaN':
        return np.nan
    else:
        duration=[]
        numero=[]
        for i in str(dato):
            duration.append(i)

        for i in duration:
            if i == " ":
                break
            else:
                numero.append(i)
        tipo=duration[len(numero)+1:len(duration)]
        resulta=''.join(tipo)
    return resulta
```

```python
# Se crean dos campos nuevos para dividir el campo duration en int y type.
df_stream_platforms['duration_int']=df_stream_platforms['duration'].map(lambda x:conversion_numero_duration(x)) 
df_stream_platforms['duration_type']=df_stream_platforms['duration'].map(lambda x:conversion_tipo_duration(x) )
```

```python
# Se reemplazan los valores faltantes.
df_stream_platforms['duration_int']= df_stream_platforms['duration_int'].replace('nan','0')
df_stream_platforms['duration_type']= df_stream_platforms['duration_type'].replace('season','seasons')
# Se cambia el tipo del campo a entero y se cambia todo en minuscula en el campo type. 
df_stream_platforms['duration_int']= df_stream_platforms['duration_int'].astype('Int32')
df_stream_platforms['duration_type'] =df_stream_platforms['duration_type'].str.lower()
```

```python
# Se exporta todo a un csv con los datos transformados.
df_stream_platforms.to_csv('stream_platforms.csv', index= False)
```
